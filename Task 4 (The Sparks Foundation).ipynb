```python
#Plotting Countries vs Attacks
data['country'].value_counts()
plt.figure(figsize=(20,8))
sns.barplot(x=data['country_txt'].value_counts()[:20].index,y=data['country_txt'].value_counts()[:20].values)
plt.title('TOP COUNTRIES ATTACKED')
plt.xticks(rotation=40)
plt.xlabel('Country')
plt.ylabel('Attacks')
plt.show()
```

```python
#Plotting Regions vs Attacks
data['region_txt'].value_counts()
plt.figure(figsize=(20,8))
sns.countplot(x='region_txt',data=data)
plt.xticks(rotation=40)
plt.title('REGIONS ATTACKED')
plt.ylabel('Attacks')
plt.xlabel('Regions')
plt.show()
```

```python
#counting the number of cities in the dataset
data['city'].value_counts()
```

```python
#Plotting Year vs Attacks
plt.subplots(figsize=(15,10))
sns.countplot(x='iyear',data=data,palette='tab10')
plt.xticks(rotation=60)
plt.title('TERROR ATTACKS YEARLY')
plt.ylabel('Attacks')
plt.xlabel('Year')
plt.show()
```

```python
#Plotting Attack Type vs Attacks
plt.figure(figsize=(20,10))
sns.countplot(x=data['attacktype1_txt'],order=data['attacktype1_txt'].value_counts().index,palette='rainbow')
plt.xticks(rotation=40)
plt.title('ATTACKS OF DIFFERENT TYPES')
plt.xlabel('Attack Type')
plt.ylabel('Attacks')
plt.show()
```

```python
#Plotting Targets vs Attacks
plt.figure(figsize=(22,10))
sns.countplot(x=data['targtype1_txt'],order=data['targtype1_txt'].value_counts().index,palette='tab10')
plt.xticks(rotation=70)
plt.title('TARGETS ATTACKED')
plt.xlabel('Targets')
plt.ylabel('Attacks')
plt.show()
```

```python
#Plotting Groups vs it's Attacks
plt.figure(figsize=(20,8))
sns.barplot(x=data['gname'].value_counts()[:20].index,y=data['gname'].value_counts()[:20].values,palette='pastel')
plt.xticks(rotation=90)
plt.title('GROUPS ATTACKED')
plt.xlabel('Group')
plt.ylabel('Attacks')
plt.show()
```

```python
#Plotting City vs it's Attacks
data['city'].value_counts()
plt.figure(figsize=(20,10))
sns.barplot(x=data['city'].value_counts()[:20].index,y=data['city'].value_counts()[:20].values, palette = "tab10")
plt.title('CITIES ATTACKED')
plt.xticks(rotation=40)
plt.xlabel('City')
plt.ylabel('Attacks')
plt.show()
```

```python
#Plotting Weapons vs it's Frequency
data['weaptype1_txt'].value_counts()
plt.figure(figsize=(20,12))
color=['lightblue','red','lightpink','brown']
sns.barplot(x=data['weaptype1_txt'].value_counts()[:20].index,y=data['weaptype1_txt'].value_counts()[:20].values , palette = "pastel")
plt.title('FREQUENCY OF WEAPONS USED')
plt.xticks(rotation=90)
plt.xlabel('Weapons')
plt.ylabel('Frequency')
plt.show()
```
